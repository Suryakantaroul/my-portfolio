Analysis


```python
# How many customers churned?
df['Churn'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
#cCalculate churn %
df['Churn'].value_counts(normalize=True) * 100
```

```python
#Churn vs Contract
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

### Insight:
Customers with month-to-month contracts have a higher churn rate compared to long-term contracts.

```python
#MonthlyCharges
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

Customers with higher monthly charges show a higher probability of churn.

```python
#Tenure vs Churn(This is one of the strongest churn indicators)
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

 #####Customers with lower tenure are more likely to churn.
 #####Long-term customers tend to stay.

```python
#Payment Method vs Churn
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
#Internet Service vs Churn
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.show()
```

```python
#Senior Citizen vs Churn
sns.countplot(x='SeniorCitizen',hue='Churn',data=df)
plt.show()
```

```python
#Tech Support vs Churn
sns.countplot(x='TechSupport', hue='Churn', data=df)
plt.show()
```

# Final Summary

## Key Insights:
- Month-to-month customers churn more
- Customers with low tenure are more likely to churn
- Higher monthly charges increase churn
- Customers using electronic check show higher churn
- Lack of tech support leads to higher churn

## Business Recommendations:
- Promote long-term contracts
- Improve onboarding for new customers
- Optimize pricing strategies
- Provide better customer support
